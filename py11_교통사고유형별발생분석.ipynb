```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

import pandas as pd
from datetime import datetime



# 폰트 설정
d2_path = './ref/D2Coding-Ver1.3.2-20180524.ttf'

fm.fontManager.addfont(d2_path)
matplotlib.rcParams["font.family"] = "D2Coding"


# 파일 load 
file_path = './ref/전국교통사고다발지역표준데이터.csv'
ori = pd.read_csv(file_path, encoding="EUC-KR")  # utf-8

ori
```

```python
ori_data = ori.groupby(by='사고유형구분')['사고건수'].sum().reset_index(name = 'count')

ori_data = ori_data.sort_values('count', ascending=False)
ori_data = ori_data.rename(columns={'사고유형구분':'type'})
ori_data
```

```python
ori_data.plot(kind='pie', 
              y='count', 
              labels=ori_data['type'], 
              figsize=(10,10), 
              startangle=0, 
              autopct='%2.1f%%', 
              title='각 사고 유형별 발생 통계', 
              cmap=matplotlib.cm.Greens)
plt.legend(loc='best')
plt.savefig('./chart/accident.jpg')
plt.show()
```

```python
ex_data = ori.copy()
ex_data = ex_data.rename(columns={'사고연도':'year', '사고유형구분':'type', '사고건수':'accident'})

# ex2_data
# ex2_data = ex_data['사고유형구분']

arr_data = list(set(ex_data['type']))

data_list = []

for t in arr_data:
    data_list.append(ex_data[ex_data['type'] == t])


data_list
```

```python
total_list = []



for d in data_list:
    result = d.groupby('year')["accident"].sum().reset_index(name='count')
    total_list.append(result)

for i in range(len(arr_data)):
    total_list[i].plot(
        kind='pie',
        y='count',
        labels=total_list[i]['year'],
        figsize=(10,10),
        startangle=90,
        autopct='%.1f%%',
        title=f'연도별 {arr_data[i]} 사고 발생 통계',
        cmap=matplotlib.cm.Blues
    )
    
    plt.savefig(f'./chart/chart{i}.png')
    plt.show()
    
```
