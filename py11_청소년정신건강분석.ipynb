```python
import pandas as pd


file_path = "./ref/청소년정신건강20250304.csv"

ori = pd.read_csv(file_path)

ori
```

```python
# 컬럼명 변경

new_name = ['시점', '스트레스', '스트레스남학생', '스트레스여학생', '우울감경험률', '우울남학생', '우울여학생', '자살생각률', '자살남학생', '자살여학생']
ori.columns = new_name
ori
```

```python
# 시점, 0번 행 삭제

ori = ori.loc[1:,'스트레스':]
ori  
```

```python
# 2번 행 추가 및 응답 열 생성

ori.loc[2] = 100 - ori.loc[1,'스트레스':].astype(float)
ori['응답'] = ['그렇다', '아니다']

ori
```

```python
# 컬럼하나를 인덱스로 지정

ori = ori.set_index('응답')
ori
```

```python
file_path2 = './ref/청소년정신건강2023.csv'

data_2023 = pd.read_csv(file_path2)

data_2023.columns = new_name

data_2023 = data_2023.loc[1:, '스트레스':]

data_2023.loc[2] = 100 - data_2023.loc[1, '스트레스':].astype(float)
data_2023['응답'] = ['그렇다', '아니다']
data_2023 = data_2023.set_index('응답')

data_2023
```

```python
fig, ax = plt.subplots(2,3, figsize=(10,8))



data_2023['스트레스'].astype(float).plot.pie(explode=[0,0.05], ax = ax[0][0], autopct='%.1f%%')
ax[0][0].set_title('스트레스를 받은 적이 있다.')
ax[0][0].set_ylabel('')

data_2023['우울감경험률'].astype(float).plot.pie(explode=[0,0.05], ax = ax[0][1], autopct='%.1f%%')
ax[0][1].set_title('우울을 경험한적 있다.')
ax[0][1].set_ylabel('')

data_2023['자살생각률'].astype(float).plot.pie(explode=[0,0.09], ax = ax[0][2], autopct='%.1f%%')
ax[0][2].set_title('자살을 고민한 적이 있다.')
ax[0][2].set_ylabel('')


ori['스트레스'].astype(float).plot.pie(explode=[0,0.05], ax = ax[1][0], autopct='%.1f%%')
ax[1][0].set_title('스트레스를 받은 적이 있다.')
ax[1][0].set_ylabel('')

ori['우울감경험률'].astype(float).plot.pie(explode=[0,0.05], ax = ax[1][1], autopct='%.1f%%')
ax[1][1].set_title('우울을 경험한적 있다.')
ax[1][1].set_ylabel('')

ori['자살생각률'].astype(float).plot.pie(explode=[0,0.09], ax = ax[1][2], autopct='%.1f%%')
ax[1][2].set_title('자살을 고민한 적이 있다.')
ax[1][2].set_ylabel('')


plt.suptitle('2023 ~ 2024년 청소년 정신건강관련 통계')
plt.show()
```

- 23년에 비해 24년에 청소년들의 스트레스 및 우울 경험이 늘어난것이 확인된다.
