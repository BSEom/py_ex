diamonds: 다이아몬드의 품질과 가격 등의 데이터를 포함한 데이터셋으로, 회귀 분석에 자주 활용됩니다.

glue: 자연어 처리(NLP) 벤치마크 데이터셋 중 하나일 가능성이 있습니다.

iris: 고전적인 머신러닝 데이터셋으로, 꽃의 종 분류 문제에 자주 사용됩니다.

mpg: 자동차의 연비(miles per gallon) 데이터를 포함한 데이터셋입니다.

penguins: 펭귄의 생물학적 특성을 다룬 데이터셋으로, 'iris'의 대안으로 종종 사용됩니다.

planets: 행성과 우주 관련 데이터를 포함할 가능성이 높습니다.

seaice: 바다 얼음 데이터와 같은 환경 관련 정보를 제공할 수 있습니다.

taxis: 택시 이동 데이터를 다룬 데이터셋으로, 이동 패턴 분석에 사용됩니다.

tips: 레스토랑 팁 데이터를 포함하며, 소규모 분석 실습에 유용합니다.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.font_manager as fm

# 리눅스용 폰트
plt.rc('font', family="NanumGothic")
# plt.text(0.2,0.4,'한글')
```

```python
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1) # 1행 2열 중에 1번째
ax2 = fig.add_subplot(1,2,2) # 1행 2열 중에 2번째

sns.stripplot(x='day', y='tip', hue='sex', data=tips, alpha=0.7, ax=ax1)
sns.swarmplot(x='day', y='tip', hue='sex', data=tips, palette='Set2', alpha=0.7, ax=ax2)

ax1.set_title('Strip plot 으로 표현한 날짜와 성별과의 팁관계')
ax2.set_title('Swarm plot 으로 표현한 날짜와 성별과의 팁관계')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1) # 1행 2열 중에 1번째
ax2 = fig.add_subplot(1,2,2) # 1행 2열 중에 2번째

sns.regplot(x='total_bill', y='tip', data=tips, color="blue", scatter_kws={"s":50, "alpha":0.5}, line_kws={"linestyle":"--", "color":"red"}, ax=ax1)
sns.regplot(x='total_bill', y='tip', data=tips, color="blue", scatter_kws={"s":50, "alpha":0.5}, fit_reg=False , ax=ax2)

fig.suptitle("Scatter plots 선형회귀의 ON / OFF 비교")
ax1.set_title('선형회귀 ON')
ax2.set_title('선형회귀 OFF')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

sns.pairplot(iris, hue="species", markers=["o","s","D"])

plt.show()
```

```python
sns.boxplot(x='species', y='sepal_length', data=iris)

plt.title("꽃받침 길이와 종의 관계")
plt.show()
```

```python
sns.violinplot(x='species', y='petal_length', data=iris, hue='species', palette='muted')

plt.title('종에 따른 꽃잎의 길이 분포')
plt.show()
```

```python
fig = plt.figure(figsize=(15,3))

ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

sns.boxplot(x='species', y='sepal_length', data=iris, ax=ax1)
sns.violinplot(x='species', y='petal_length', data=iris, hue='species', palette='muted', ax=ax2)
sns.histplot(x='petal_length', data=iris, hue='species', kde=True, element='step', palette='Set2', ax=ax3)


ax1.set_title("꽃받침 길이와 종의 관계")
ax2.set_title('종에 따른 꽃잎의 길이 분포1')
ax3.set_title('종에 따른 꽃잎의 길이 분포2')

sns.lmplot(x='sepal_length', y='sepal_width', data=iris, hue='species', markers=["o","s","D"], height=6, aspect=1.2)
plt.title('꽃받침 길이와 너비의 관계')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

exercise = sns.load_dataset('exercise')

exercise.sample(5)
```

```python
# point plot

plt.figure(figsize=(10,6))
sns.pointplot(x='time', y='pulse', hue='kind', data=exercise, capsize=0.1)

plt.title('심박수와 시간별 운동종류와의 관계')
plt.show()
```

```python
# box plot

plt.figure(figsize=(7,4))
sns.boxplot(x='time', y='pulse', hue='kind', data=exercise, palette='Set2')

plt.title('심박수와 시간별 운동종류와의 관계')
plt.show()
```

```python
# swarm plot

plt.figure(figsize=(7,4))
sns.swarmplot(x='time', y='pulse', hue='kind', data=exercise, dodge=True, palette='Set1')

plt.title('시간과 운동 유형에 따른 개별 맥박 측정')
plt.show()
```

이 그래프는 시간 경과에 따른 운동 유형별 맥박 변화를 보여주며, 운동 강도가 높을수록 맥박이 증가하는 경향이 뚜렷하게 나타난다. 휴식 상태에서는 시간이 지나도 맥박 변화가 거의 없지만, 걷기(walking)는 점진적인 증가를 보이며, 달리기(running)는 가장 큰 폭으로 맥박이 상승하여 30분 후 최고 수준에 도달한다. 이는 운동 강도가 높을수록 심박수가 더 빠르게 증가하며, 지속적인 운동이 심혈관 시스템에 미치는 영향을 반영하는 결과로 해석할 수 있다.
